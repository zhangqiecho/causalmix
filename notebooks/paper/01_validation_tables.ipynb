## Validation tables (LaTeX)

```python
from pathlib import Path
import sys

# --- Resolve repo root (works whether you run from repo root or within notebooks/) ---
REPO_ROOT = Path.cwd().resolve()
if not (REPO_ROOT / "src").exists():
    if (REPO_ROOT.parent / "src").exists():
        REPO_ROOT = REPO_ROOT.parent
    elif (REPO_ROOT.parent.parent / "src").exists():
        REPO_ROOT = REPO_ROOT.parent.parent

sys.path.insert(0, str(REPO_ROOT))
sys.path.insert(0, str(REPO_ROOT / "src"))

from IPython.display import display
from notebooks.set_up import (
    ensure_dirs,
    TABLE_DIR, 
    VALIDATION_DIR
)

ensure_dirs()

# Inputs
output_result = str(VALIDATION_DIR) + "/"

# Outputs
output_table = TABLE_DIR
output_table = str(output_table) + "/"
```

# Get latex tables for the validation part

```python
# Purpose: create latex files for all tables in the results sections
import pandas as pd
file_names = ["m1_tables_mean.xlsx", "m2_tables_mean.xlsx", "m3_tables_mean.xlsx"]

excel_tables = [pd.read_excel(output_result + fn, sheet_name=None, engine="openpyxl") for fn in file_names]

m1_dist, m1_privacy, m1_causal = [excel_tables[0][list(excel_tables[0].keys())[i]] for i in range(3)]
m2_dist, m2_privacy, m2_causal = [excel_tables[1][list(excel_tables[1].keys())[i]] for i in range(3)]
m3_dist, m3_privacy, m3_causal = [excel_tables[2][list(excel_tables[2].keys())[i]] for i in range(3)]

m1_privacy = m1_privacy.drop(columns=["Category"])
m2_privacy = m2_privacy.drop(columns=["Category"])
m3_privacy = m3_privacy.drop(columns=["Category"])

display(m1_privacy)
display(m2_privacy)
display(m3_privacy)
```

```python
# save tex files for all three tables
latex = m1_dist.to_latex(
    index=False,
    escape=False,
    column_format="llrrr"
)

with open(output_table + "m1_dist.tex", "w", encoding="utf-8") as f:
    f.write(latex)

latex = m1_privacy.to_latex(
    index=False,
    escape=False,
    column_format="llrr"
)

with open(output_table + "m1_privacy.tex", "w", encoding="utf-8") as f:
    f.write(latex)

latex = m1_causal.to_latex(
    index=False,
    escape=False,
    column_format="lllrr"
)

with open(output_table + "m1_causal.tex", "w", encoding="utf-8") as f:
    f.write(latex)

latex = m2_dist.to_latex(
    index=False,
    escape=False,
    column_format="llrrr"
)

with open(output_table + "m2_dist.tex", "w", encoding="utf-8") as f:
    f.write(latex)

latex = m2_privacy.to_latex(
    index=False,
    escape=False,
    column_format="llrr"
)

with open(output_table + "m2_privacy.tex", "w", encoding="utf-8") as f:
    f.write(latex)

latex = m2_causal.to_latex(
    index=False,
    escape=False,
    column_format="lllrr"
)

with open(output_table + "m2_causal.tex", "w", encoding="utf-8") as f:
    f.write(latex)

latex = m3_dist.to_latex(
    index=False,
    escape=False,
    column_format="llrrr"
)

with open(output_table + "m3_dist.tex", "w", encoding="utf-8") as f:
    f.write(latex)

latex = m3_privacy.to_latex(
    index=False,
    escape=False,
    column_format="llrr"
)

with open(output_table + "m3_privacy.tex", "w", encoding="utf-8") as f:
    f.write(latex)

latex = m3_causal.to_latex(
    index=False,
    escape=False,
    column_format="lllrr"
)

with open(output_table + "m3_causal.tex", "w", encoding="utf-8") as f:
    f.write(latex)
```
